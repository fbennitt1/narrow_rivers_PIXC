```python
import os

import contextily as ctx
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr

from pandarallel import pandarallel
from matplotlib import colors
from reaches import *
```

### Pixel Cloud

#### Original PIXC granule

```python
data_path = '/nas/cee-water/cjgleason/fiona/data/small_rivers/mar_2024_ver_c/'
pixc_path = os.path.join(data_path, 'SWOT_L2_HR_PIXC_012_298_080L_20240317T084237_20240317T084248_PIC0_01.nc')
tile_name = pixc_path[-71:-3]

# Read in xarray
ds_PIXC = xr.open_mfdataset(paths=pixc_path, group = 'pixel_cloud', engine='h5netcdf')
```

```python
# Make dict for legend labels
flags = ds_PIXC.classification.flag_meanings.split() # extract each flag meaning
codes = {idx:k for idx, k in enumerate(flags, start=1)}
```

#### Prepped pseudo-pixels

```python
data_path = '/nas/cee-water/cjgleason/fiona/narrow_rivers_PIXC_data/PIXC_v2_0_pseudo_pixels_filtered/'

pixc_path = os.path.join(data_path, 'SWOT_L2_HR_PIXC_012_298_080L_20240317T084237_20240317T084248_PIC0_01.parquet')

file_name = pixc_path[-76:-8]

pass_num = int(file_name[20:23])
```

```python
gdf_PIXC = gpd.read_parquet(pixc_path)
```

```python
# Get single pixel for selecting correct nadir segment
pixel_pt = gdf_PIXC.iloc[0].geometry
```

```python
# Find correct nadir segment and return its geometry
nadir_segment_ln = findNadir(pass_num=pass_num, pixel_pt=pixel_pt)
```

### Make pseudo pixels

```python
# Set along-track pixel resolution
azimuth_res = 21 # meters
```

```python
pandarallel.initialize(nb_workers=int(os.environ.get('SLURM_CPUS_PER_TASK')))
```

```python
# Make pseudo pixels
gdf_PIXC['pseudo_pixel'] = gdf_PIXC.parallel_apply(user_defined_function=makePseudoPixels,
                                                         args=(nadir_segment_ln,
                                                               azimuth_res),
                                                         axis=1)
```

```python
pseudo = gdf_PIXC.drop(columns='geometry').set_geometry('pseudo_pixel')
```

### NHDPlus HR

```python
i = 4 # HUC4 0108, Connecticut
```

```python
basin, huc2, huc4 = readNHD(index=i)
```

```python
basin = basin.to_crs('EPSG:32618')
```

```python
buffer = basin.copy()
```

```python
# Buffer reach by 1/2 channel width
buffer['geometry'] = buffer['geometry'].buffer(distance=(buffer['WidthM_Max']/2), cap_style='flat')
```

### PIXC plots

```python
palette = {3: 'cornflowerblue', 4: 'blue', 5: 'hotpink',
           6: 'darkkhaki', 7: 'springgreen'} # 6: 'lightsalmon', 7: 'lightblue'
```

```python
# Connecticut
bounds = [694000, 4688500, 699000, 4695166] # 5 km x 6.67 km

PIXC_c = gdf_PIXC.clip(mask=bounds)
buffer_c = buffer.clip(mask=bounds)

nums = list(np.unique(PIXC_c.klass).astype('int'))
labels = [codes[x] for x in nums]

# Make cmap
cmap = colors.ListedColormap([palette[x] for x in nums])

fig, ax = plt.subplots(figsize=(5,6.6667))
# basin.clip(mask=bound_box).plot(ax=ax, alpha=0.5)
buffer_c.plot(ax=ax, alpha=0.7, color='k')
PIXC_c.plot(ax=ax, alpha=0.9, column=PIXC_c.klass,
              categorical=True, cmap=cmap, s=1)

# Basemap
ctx.add_basemap(ax, crs=gdf_PIXC.crs, source=ctx.providers.CartoDB.Positron, alpha=0.9)

plt.xlim(694000, 699000) # 5 km x 5 km
plt.ylim(4688500, 4695166) # 5 km x 6.6 km

plt.axis('off')
# plt.savefig(fname=save_path + 'pixc_conn.png', dpi=300, bbox_inches='tight')
```
