```python
def filterVersion_riverSP(directories, output_directory):
    """
    Reads in all filenames, sorts them, and retrieves the best version
    of each file (e.g. PGC0 over PIC0, PIC0_03 over PIC0_02). Copies
    all filtered files with the same base name (regardless of extension)
    from the input directories to a new directory.

    Parameters:
    directories (list): List of directories to search for .shp files and related files.
    output_directory (str): Path to the directory where filtered files will be copied.
    
    Returns:
    None
    """

    # List to store all .shp file paths
    files = []
    
    # Get all .shp file names from directories
    for directory in directories:
        for file in os.listdir(directory):
            if file.endswith(".shp"):
                files.append(file)
    
    print(f"There are {str(len(files))} original files in directory")
    
    # Make DataFrame of filenames
    granules = pd.DataFrame({'files': files})
    granules['cycle'] = granules['files'].str.slice(25, 28)
    granules['pass'] = granules['files'].str.slice(29, 32)
    granules['version'] = granules['files'].str.slice(-11, -7)
    granules['counter'] = granules['files'].str.slice(-6, -4)    
    granules['region'] = granules['files'].str.slice(33, 35)    

    # Sort the files
    granules = granules.sort_values(by=['cycle', 'pass', 'version', 'counter'],
                                    ascending=[True, True, True, False])    

    # Keep only the best version of each granule
    granules = granules.drop_duplicates(subset=['cycle', 'pass'],
                                        keep='first')    

    # Extract the file names of files passing the test
    keep_files = list(granules['files'])
    
    #########################################
    #Create output directory with similar structure to original
    output_directory = os.path.join(output_directory,
                                    granules['region'].iloc[0], 'Reach')

    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    #########################################

    # Extract base names (file name without extensions) from the list
    base_names = set(os.path.splitext(os.path.basename(file))[0] for file in keep_files)

    # Loop over the directories to find all files with matching base names
    for directory in directories:
        for file in os.listdir(directory):
            # Get the file's base name
            base_name, extension = os.path.splitext(os.path.basename(file))

            # If the base name is in the list of files to keep, copy all files with that base name
            if base_name in base_names:
                full_path = os.path.join(directory, file)
                destination = os.path.join(output_directory, os.path.basename(file))

                # Perform the copy operation
                shutil.copy(full_path, destination)
                #print(f"Copied: {full_path} to {destination}")

        print(f"Copied the unique and most recently processed {str(len(keep_files))} files to {output_directory}")
```

#### For PIXC

```python
def filterVersion_PIXC(directories, output_directory):
    """
    Reads in all filenames, sorts them, and retrieves the best version
    of each file (e.g. PGC0 over PIC0, PIC0_03 over PIC0_02). Copies
    all filtered files from the input directories to a new directory.

    Parameters:
    directories (list): List of directories to search for best files.
    output_directory (str): Path to the directory where filtered files will be copied.
    
    Returns:
    None
    """

    # List to store all .shp file paths
    files = []
    
    # Get all file names from directories
    for directory in directories:
        for file in os.listdir(directory):
            files.append(file)
    
    print(f"There are {str(len(files))} original files in directory.")
    
    # Make DataFrame of filenames
    granules = pd.DataFrame({'files': files})
    granules['cycle'] = granules['files'].str.slice(25, 28)
    granules['pass'] = granules['files'].str.slice(29, 32)
    granules['tile'] = granules['files'].str.slice(24,28)
    granules['version'] = granules['files'].str.slice(-11, -7)
    granules['counter'] = granules['files'].str.slice(-6, -4)    

    # Sort the files
    granules = granules.sort_values(by=['cycle', 'pass', 'tile', 'version', 'counter'],
                                    ascending=[True, True, True, True, False])    

    # Keep only the best version of each granule
    granules = granules.drop_duplicates(subset=['cycle', 'pass', 'tile'],
                                        keep='first')    

    # Extract the file names of files passing the test
    keep_files = list(granules['files'])
    
    #########################################
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    #########################################

    # Loop over the directories to find all files with matching base names
    for directory in directories:
        for file in os.listdir(directory):

            # If filename is in keep_files, copy file to new dir
            if file in keep_files:
                full_path = os.path.join(directory, file)
                destination = os.path.join(output_directory,
                                           os.path.basename(file))

                # Perform the copy operation
                shutil.copy(full_path, destination)
                #print(f"Copied: {full_path} to {destination}")

        print(f"Copied the unique and most recently processed {str(len(keep_files))} files to {output_directory}")
```
